```python
import pandas as pd
import numpy as np
import os
```

```python
raw_data_path=os.path.join(os.path.pardir,'data','raw')
train_file_path=os.path.join(raw_data_path,'train.csv')
test_file_path=os.path.join(raw_data_path,'test.csv')
```

```python
#read the data with all default params
train_df=pd.read_csv(train_file_path,index_col='PassengerId')
test_df=pd.read_csv(test_file_path,index_col='PassengerId')
```

```python
test_df['Survived']=-888 # Adding Survived with a default value
```

```python
df=pd.concat((train_df,test_df),axis=0) #Concatinating both data frames, 0 for row, 1 for col
```

```python
# use & or | operators to build complex logic
male_passengers_first_class=df.loc[(df.Sex=='male')&(df.Pclass==1),:]
print('Number of male passengers in first class: {0}'.format(len(male_passengers_first_class)))
male_passengers_first_class
```

```python
df.describe() #provide summary statistics
```

```python
print('Mean fare:{0}'.format(df.Fare.mean()))
print('Median fare:{0}'.format(df.Fare.median()))
```

```python
print('Min fare:{0}'.format(df.Fare.min()))
print('Max fare:{0}'.format(df.Fare.max()))
print('Range fare:{0}'.format(df.Fare.max() - df.Fare.min()))
print('1st Quartile fare:{0}'.format(df.Fare.quantile(.25)))
print('2nd Quartile fare:{0}'.format(df.Fare.quantile(.5)))
print('3rd Quartile fare:{0}'.format(df.Fare.quantile(.75)))
print('Variance fare:{0}'.format(df.Fare.var()))
print('Std fare:{0}'.format(df.Fare.std()))
```
